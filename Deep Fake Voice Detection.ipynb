```python
import sklearn
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Deep Fake Dataset.csv")
column_list = df.columns.tolist()
```

```python
df
```

```python
# Get Range of all variables
range_list = []
for column_name in column_list[:-1]:
    col = df[column_name].to_list()
    range_list.append((column_name, max(col) - min(col)))

range_list.sort(key = lambda x: x[1])
range_list
```

```python
ss = StandardScaler()

data = df.to_numpy()
np.random.shuffle(data)

train_split = data.shape[0] * 6 // 10
valid_split = data.shape[0] * 8 // 10

train_data = data[: train_split, :]
valid_data = data[train_split : valid_split, :]
test_data = data[valid_split :, :]

print(train_data.shape)
print(valid_data.shape)
print(test_data.shape)

train_X, train_y = ss.fit_transform(train_data[:, :-1]), train_data[:, -1]
valid_X, valid_y = ss.transform(valid_data[:, :-1]), valid_data[:, -1]
test_X, valid_y = ss.transform(test_data[:, :-1]), test_data[:, -1]

# Get Range of all variables
range_list = []
for i, column_name in enumerate(column_list[:-1]):
    col = train_X[:, i]
    range_list.append((column_name, max(col) - min(col)))

range_list.sort(key = lambda x: x[1])
range_list
```

```python
pca = PCA(n_components=8)
pca.fit(train_X)

sum(pca.explained_variance_ratio_)
```

```python
train_X = pca.transform(train_X)
valid_X = pca.transform(valid_X)
test_X = pca.transform(test_X)

print(f"Train X shape: {train_X.shape}")
print(f"Valid X shape: {valid_X.shape}")
print(f"Test X shape: {test_X.shape}")
```
